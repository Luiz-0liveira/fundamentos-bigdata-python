Exercício 04 – Criando relatórios em função do arquivo dados-vendas.xlsx.

a) Crie um dataframe, utilizando o arquivo dados-vendas.xlsx. Crie data frames
para dividir a planilha vendas em 4 arquivos. São eles:

Tabela Fato

- ID-Produto;
- ID-Loja;
- ID-Vendedor;
- Data;
- ValorVenda.

Tabela Produtos

- ID-Produto;
- Produto;
- Categoria;
- Segmento;
- Fabricante;

Tabela Lojas

- ID-Loja;
- Cidade;
- Estado

Tabelas Vendedor.

- ID-Vendedor;
- Vendedores;

```python
# Importar a biblioteca do pandas
import pandas as pd

# Mapear o local do documento fonte

caminho = 'C:\\Users\\FIC\\Documents\\Luiz Antonio Python\\003 - Arquivos\\'
arquivo = 'dados-vendas.xlsx'

# Atribuir os dados deste arquivo, em um dataframe

base = pd.read_excel(caminho + arquivo, sheet_name='Vendas')

# Exibir os cabeçalhos das colunas

base.head()
```

```python
# Criando a tabela fato comas colunas:


base[['ID-Produto','Loja','ID-Vendedor','Data Venda','ValorVenda']]
```

```python

# Criando a tabela fato comas colunas:
df_dados = base[['ID-Produto','Loja','ID-Vendedor','Data Venda','ValorVenda']]

# Renomear a coluna Loja

df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

df_dados.head()
```

```python
# salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'

arquivo = 'tabela_fato.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```

##  Tabela Produtos

```python

# Criando a Tabela Produtos com as colunas:
df_dados = base[['ID-Produto','Produto','Categoria','Segmento','Fabricante']]

# Renomear a coluna Loja

df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

df_dados.head()
```

```python
# salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'

arquivo = 'tabela produtos.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```

### Tabela Lojas

```python
# Criando a Tabela Produtos com as colunas:
df_dados = base[['Loja','Cidade','Estado']]

# Renomear coluna

df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

# Remover Duplicadas

df_dados = df_dados.drop_duplicates()

df_dados.head()
```

```python
# salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'

arquivo = 'tabela_lojas.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```

### Tabelas Vendedor

```python
# Criando a Tabela Produtos com as colunas:
df_dados = base[['ID-Vendedor','Vendedor']]

# Renomear a coluna Loja

df_dados = df_dados.rename(columns={'Loja':'ID-Loja'})

df_dados.head()
```

```python
# salvar o arquivo na pasta dados-vendas

caminho_novo = caminho + 'dados-vendas\\'

arquivo = 'tabela_vendedor.xlsx'

df_dados.to_excel(caminho_novo + arquivo, index= False)
```
