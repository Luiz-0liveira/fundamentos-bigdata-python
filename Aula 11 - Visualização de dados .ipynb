### Importando as bibliotecas

Matplotlib é a principal biblioteca para geração de gráficos em Python. A biblioteca possui uma
infinidade de possibilidades, sendo possível criarmos desde gráficos simples até outros mais
elaborados.

```python
import matplotlib.pyplot as plt # permite geração de gráficos
import numpy as np              # Permite trabalhar com matrizes e vetores
```

#### Primeiro Gráfico com Matplotlib

```python
#  Criando os vetores (conjuntode dados)

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Plotar o gráfico
plt.plot(x, y)
```

### Modificando o tipo dos marcadores 

```python
plt.plot(x, y, marker= '*')
```

### Alterandoas cores das linhas

```python
# Pode usar palheta de cores do Adobe color

plt.plot(x, y, marker = '*', c='#F8115E')   # b=cor azul # g= cor verde
```

## Mudando com os pontos são ligados e o tipo de linha

```python
plt.plot(x,y,      # Dados do gráfico
    marker = '*',    # Marcadores
    c = 'red',     # Cor  da linha
    ds = 'steps',  # Forma de conexação dos pontos
    ls = '--'     # Estilo de linha

)
```

## Mudando a cor dos pontos

```python
plt.plot (x,y, marker = '*', c = 'red', ds = 'steps', ls = '--', mec =
'blue')  #
```

### Matplotlib – adicionando outros elementos

Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outras
possibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando a
atenção do que realmente importa.

```python

plt.plot(x,y)  # Adiciona dados ao gáfico
plt.xlabel('Valores de X', fontsize = 12) # Adicionar titulo ao eixo x
plt.ylabel('Valores de Y', fontsize = 12) # Adicionar título ao eixo Y
plt.title ('Título do gráfico', fontsize = 15, c='purple') # Adicionar título ao gráfico
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'orange')
```

### Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

# retornar  (Linspace) valores igualmente espaçadas em um intervalo 


x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

### Matplotlib – Múltiplos subgráficos
Dois gráficos na vertical (linha)

```python
# O comando plt.subplot permite criar grupos de gráfico
# comando plt.subplot (linhas, eixo compartilhado)

plt.subplot(2,1,1) # Primeiro gráfico 
plt.plot(x,y)       # Conjunto de dados do primeiro
plt.title('subplots')

plt.subplot(2,1,2)   # Segundo gráfico - matrix 2x2 (linha e coluna) (2,1,2) linha,coluna e elemento
plt.plot(x,z)        # Conjunto de dados do segundo gráfico
```

```python
plt.subplot(1,2,1)  # uma linha, duas colunas e um elemento 
plt.plot(x,y)
plt.title('subplots')

plt.subplot(1,2,2)
plt.plot(x,z)
```

###  Seaborn – Gráficos mais elaborados
Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos,
além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.


```python
# O ALIAS (apelido) da biblioteca SEABORN FICA COMO SNS 
# importando a biblioteca

import seaborn as sns
import numpy as np

# Retorna (Linspace) valores igualmente espaçadas em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
```

## Seaborn – Datasets embutidos
O Seaborn possui diversos dados embutidos que podem ser utilizados para estudar a própria
biblioteca. Para sabermos quais os datasets estão disponíveis, podemos utilizar a função
get_datasets_names ().

```python
sns.get_dataset_names()
# CADA UM É UMA TABELA
```

## Para acessar o dataset escolhido, usamos a função load_dataset ().
Criando um dataframe com o dataset íris.

```python
iris = sns.load_dataset('iris')
iris.head()
```

## Criando um dataframe com o dataset planets.


```python
# O DATASET PLANETS OU IRIS JÁ ESTÃO INTEGRADOS NO SEABORN
# dataset para estudo ou simulação

planetas = sns.load_dataset('planets')
planetas.head()
```

## Seaborn – Ajustando os gráficos

O Seaborn possui a função set () que permite ajustar estilos específicos para os gráficos, sem a
necessidade de ajustar manualmente todos os argumentos.

Gráfico com a configuração padrão do Matplotlib

```python
import matplotlib.pyplot as pl

sns.set()
plt.plot (x,y)
```

```python
plt.plot (x,y)
```

 O estilo default do Seaborn é o darkgrid, mas há ainda os estilos whitegrid, dark, white e ticks que
podem ser escolhidos com a função set_style ().

```python
# Alterando o tema para claro

sns.set_style('whitegrid')
plt.plot(x,y)
```

```python
# Alterando tema escuro

sns.set_style('dark')
plt.plot(x,y)
```

Os gráficos vistos até o momento são apenas com dados numéricos. E os dados categóricos? Como

Os gráficos stripplot () e swarmplot (), são gráficos em que analisamos a dispersão de uma variável
