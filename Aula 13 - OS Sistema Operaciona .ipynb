```python

# Importar as bibliotecas
import pandas as pd # manipular os dados
import os # gerenciar arquivos do sistema operacional


# Criar variavel para armazenar o caminho do diretório 
dir_path ='C:\\Users\\FIC\\Documents\\Luiz Antonio Python\\003 - Arquivos\\filiais\\'

# Criar uma lisat vazia para armazenar os caminhos completos
res = []

for path in os.listdir(dir_path):
    # Veridficar se é um arquivo válido
    if os.path.isfile(dir_path + path): 
        res.append(dir_path + path)  
#res Lista com caminhos válidos 

#Criar um dataframe vazio
file_combined = pd.DataFrame()

# Estrutura para combinar os arquivos
for file in res:
    df1 = pd.read_excel(file)
    file_combined = pd.concat([file_combined,df1])

dir_final = 'C:\\Users\\FIC\\Documents\\Luiz Antonio Python\\Exercicio 05 Lista 05\\'
fileName = 'ArquivoCombinado.xlsx'

file_combined.to_excel(dir_final + fileName)
```
