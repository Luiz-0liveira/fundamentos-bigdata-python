```python
dir_path = 'C:\\Users\\FIC\\Documents\\Luiz Antonio Python\\Desafio\\arquivos\\vendas\\'
# Criar uma lisat vazia para armazenar os caminhos completos
res = []

for path in os.listdir(dir_path):
    # Veridficar se é um arquivo válido
    if os.path.isfile(dir_path + path): 
        res.append(dir_path + path)  
#res Lista com caminhos válidos 

#Criar um dataframe vazio
file_combined = pd.DataFrame()

# Estrutura para combinar os arquivos
for file in res:
    df1 = pd.read_excel(file)
    file_combined = pd.concat([file_combined,df1])
```
