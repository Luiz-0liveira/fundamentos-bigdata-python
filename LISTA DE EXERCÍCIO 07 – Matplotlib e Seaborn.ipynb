Exercício 01 - Escreva um programa Python para desenhar uma linha com rótulo adequado no eixo x, eixo y e um título.

```python
# Importa a biblioteca

import matplotlib.pyplot as plt

# Gerar os dados para inserir no gráfico

x = range(1,50)
y = [value * 3 for value in x]

print('Valores de X')
print(range(1,50))
print('Valores de Y(Três vezes X)')
print(y)
```

```python
# Plotar um gráfico de linha

plt.plot(x, y)

# Adicionar um título ao gráfico

plt.title('Draw a line', fontsize = 15, c='blue')

# Adicionar títulos aos eixos

plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
```
